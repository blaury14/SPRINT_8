Se analizarán los datos de visitas, registros de pedidos y costos de marketing para informar las decisiones sobre la inversión en marketing.*

## Carga y Optimizacion de Datos:

### Carga de Librerias y Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar datos
visits = pd.read_csv('/datasets/visits_log_us.csv')
orders = pd.read_csv('/datasets/orders_log_us.csv')
costs = pd.read_csv('/datasets/costs_us.csv')

# Revisar los primeros registros y tipos de datos
display(visits.head())
display(orders.head())
display(costs.head())

visits.info()
orders.info()
costs.info()
```

#### ¿Cuántas personas lo usan cada día, semana y mes?

```python
# Determinar el mes de la primera actividad para cada usuario para establecer la cohorte de cada uno
visits['Cohort'] = visits.groupby('uid')['Start Ts'].transform('min').dt.to_period('M')
# Convertir la fecha de cada visita al período mensual correspondiente
visits['Order Period'] = visits['Start Ts'].dt.to_period('M')

# Crear la data de cohortes
cohort_data = visits.groupby(['Cohort', 'Order Period']).agg(n_customers=('uid', 'nunique')).reset_index()

# Crear el campo 'Period Number' como la diferencia en meses entre Order Period y Cohort
cohort_data['Period Number'] = (cohort_data['Order Period'].dt.start_time - cohort_data['Cohort'].dt.start_time).dt.days // 30

# Crear la tabla pivot para la matriz de retención
cohort_pivot = cohort_data.pivot_table(index='Cohort', columns='Period Number', values='n_customers')

# Dividir cada columna por el total de usuarios en el periodo inicial (mes 0 de cada cohorte)
cohort_size = cohort_pivot.iloc[:, 0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0)

# Visualizando la matriz de retención como un mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(retention_matrix, annot=True, fmt='.0%', cmap='Blues')
plt.title('Análisis de Cohorte - Tasas de Retención')
plt.ylabel('Mes de la Cohorte')
plt.xlabel('Meses desde la Primera Actividad')
plt.show()
```

#### ¿Con qué frecuencia los usuarios y las usuarias regresan?

```python
# Encuentra la primera visita para cada usuario
first_visits = visits.groupby('uid')['Start Ts'].min().reset_index()
first_visits.rename(columns={'Start Ts': 'First Visit Date'}, inplace=True)

# Une las primeras visitas al DataFrame de visitas original para tener la fecha de la primera visita
visits = visits.merge(first_visits, on='uid')

# Calcula si la sesión es una sesión de retorno (es decir, cualquier visita después de la primera)
visits['Is Returning'] = visits['Start Ts'] > visits['First Visit Date']

# Calcula la frecuencia de visitas de retorno
returning_users_frequency = visits['Is Returning'].mean()

# Imprimir el porcentaje de sesiones de retorno
print(f"El porcentaje de sesiones de retorno es: {returning_users_frequency * 100:.2f}%")

# Asignar el período de cohorte a cada fila
visits['Cohort Period'] = visits.apply(lambda x: (x['Start Ts'].to_period('M') - x['Cohort']).n, axis=1)

# Agrupar por Cohorte y Cohort Period, y contar usuarios únicos
cohort_data = visits.groupby(['Cohort', 'Cohort Period']).agg(n_users=('uid', 'nunique')).reset_index()

# Pivotar los datos para obtener el número de usuarios únicos en cada Cohort Period para cada Cohorte
cohort_counts = cohort_data.pivot_table(index='Cohort', columns='Cohort Period', values='n_users')

# Seleccionar la primera columna, que es el tamaño total de cada cohorte
cohort_sizes = cohort_counts.iloc[:,0]

# Dividir todos los valores de la tabla pivot por los tamaños de cohorte correspondientes para obtener la tasa de retención
retention = cohort_counts.divide(cohort_sizes, axis=0)

# Visualizar la tasa de retención con un heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(retention, annot=True, fmt='.0%', cmap='Blues')
plt.title('Análisis de Cohorte: Tasas de Retención de Usuarios')
plt.ylabel('Cohorte (Mes de Primera Actividad)')
plt.xlabel('Período desde la Primera Actividad (meses)')
plt.show()
```

#### ¿Cuándo la gente empieza a comprar? 

```python
# Convertir las fechas de string a datetime si aún no lo están
visits['Start Ts'] = pd.to_datetime(visits['Start Ts'])
orders['buy_ts'] = pd.to_datetime(orders['buy_ts'])

# 1. Determinar la fecha de la primera visita para cada usuario
visits['Fecha de Primera Visita'] = visits.groupby('uid')['Start Ts'].transform('min')

# 2. Determinar la fecha de la primera compra para cada usuario
orders['Fecha de Primera Compra'] = orders.groupby('uid')['buy_ts'].transform('min')

# 3. Unir los dos DataFrames para tener las fechas de la primera visita y compra juntas
conversion = pd.merge(visits[['uid', 'Fecha de Primera Visita']], orders[['uid', 'Fecha de Primera Compra']], on='uid', how='inner')

# 4. Calcular la diferencia en días entre la fecha de la primera visita y la primera compra
conversion['Días para la Conversión'] = (conversion['Fecha de Primera Compra'] - conversion['Fecha de Primera Visita']).dt.days

# 5. Crear categorías basadas en rangos de días
def categorizar_dias(x):
    if x == 0:
        return '0 días'
    elif 1 <= x <= 30:
        return '1-30 días'
    elif 31 <= x <= 60:
        return '31-60 días'
    elif 61 <= x <= 90:
        return '61-90 días'
    elif 91 <= x <= 120:
        return '91-120 días'
    elif 121 <= x <= 150:
        return '121-150 días'
    elif 151 <= x <= 180:
        return '151-180 días'
    elif 181 <= x <= 210:
        return '181-210 días'
    elif 211 <= x <= 240:
        return '211-240 días'
    elif 241 <= x <= 270:
        return '241-270 días'
    elif 271 <= x <= 300:
        return '271-300 días'
    elif 301 <= x <= 330:
        return '301-330 días'
    elif 331 <= x <= 360:
        return '331-360 días'
    else:
        return '361+ días'

conversion['Categoría de Conversión'] = conversion['Días para la Conversión'].apply(categorizar_dias)

# Agrupar por cohorte y categoría de conversión y contar el número de usuarios
conversion['Cohorte'] = conversion['Fecha de Primera Visita'].dt.to_period('M')
resumen_conversión = conversion.groupby(['Cohorte', 'Categoría de Conversión']).size().unstack(fill_value=0)

# Reordenar las columnas para que estén en orden cronológico
orden = [
    '0 días', '1-30 días', '31-60 días', '61-90 días', '91-120 días', 
    '121-150 días', '151-180 días', '181-210 días', '211-240 días',
    '241-270 días', '271-300 días', '301-330 días', '331-360 días', '361+ días'
]
resumen_conversión = resumen_conversión[orden]

# Mostrar el DataFrame organizado
display(resumen_conversión)
```

**R:** 
Al analizar la tabla, podemos sacar algunas conclusiones importantes sobre cómo las visitas se convierten en compras a lo largo del tiempo, desde `junio de 2017 hasta mayo de 2018.` Aquí dejo algunas observaciones:

1. **Conversión Rápida:** Muchos usuarios compran pronto después de su primera visita, especialmente dentro de los primeros 30 días. Esto resalta la importancia de captar la atención y convertir a los nuevos visitantes en este período inicial.
2. **Menos Compras con el Tiempo:** A medida que pasa más tiempo desde la primera visita, vemos menos conversiones. Esto sugiere que si los visitantes no compran inicialmente, es menos probable que lo hagan en el futuro.
3. **Posible Influencia Estacional:** Algunos cambios en las conversiones podrían estar relacionados con eventos o promociones específicas. Por ejemplo, un aumento repentino en las compras podría estar vinculado a una campaña de marketing exitosa.
4. **Cohortes de 2018 con Menor Enganche:** Las cohortes más recientes muestran una tendencia a tener menos conversiones a lo largo del tiempo, lo que podría indicar cambios en las estrategias de marketing o en la competencia del mercado.
5. **Descenso a Largo Plazo:** En general, vemos una disminución en la retención y conversión a largo plazo, especialmente en cohortes más recientes. Esto destaca la importancia de mantener el interés del cliente y mejorar la lealtad a lo largo del tiempo.

#### ¿Cuántos pedidos hacen durante un período de tiempo dado?

```python
# Convertir las fechas de compra a un período mensual
orders['Mes'] = orders['buy_ts'].dt.to_period('M')

# Contar cuántos pedidos hace cada usuario por mes
order_counts = orders.groupby(['uid', 'Mes']).size().reset_index(name='Cantidad de Pedidos')

# Agrupar por mes y sumar los pedidos
monthly_orders = order_counts.groupby('Mes')['Cantidad de Pedidos'].sum().sort_index()
display(monthly_orders.describe())

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
monthly_orders.plot(kind='bar', color='skyblue')
plt.title('Número Total de Pedidos por Mes')
plt.xlabel('Mes')
plt.ylabel('Número Total de Pedidos')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

#### ¿Cuánto dinero traen? (LTV)

```python
# Calcula la fecha de la primera compra para cada usuario
orders['First Purchase Date'] = orders.groupby('uid')['buy_ts'].transform('min')

# Convierte la fecha de la primera compra en una columna de tipo 'periodo' agrupado por mes
orders['Cohort'] = orders['First Purchase Date'].dt.to_period('M')

# Calcula la 'edad' en meses para cada compra
orders['age_month'] = (orders['buy_ts'].dt.to_period('M') - orders['Cohort']).apply(lambda r: r.n)

# Ahora agrupa por 'Cohort' y 'age_month' para calcular el LTV
ltv = orders.groupby(['Cohort', 'age_month']).agg({'revenue': 'sum'}).reset_index()

# Calcula el LTV acumulado por cohorte
ltv['LTV'] = ltv.groupby('Cohort')['revenue'].cumsum() / orders.groupby('Cohort')['uid'].nunique().reindex(ltv['Cohort'].values).values

# Pivotea el DataFrame para el mapa de calor
ltv_pivot = ltv.pivot(index='Cohort', columns='age_month', values='LTV')

# Configura el tamaño de la figura
plt.figure(figsize=(15, 8))

# Crea el mapa de calor
sns.heatmap(ltv_pivot, annot=True, fmt=".2f", cmap='Blues', linewidths=.5)

# Título y etiquetas
plt.title('LTV por Cohorte de Usuario y Edad en Meses')
plt.xlabel('Edad en Meses')
plt.ylabel('Mes de Primera Orden')

# Muestra el gráfico
plt.show()
```

**R:** Observando el heatmap de LTV (Lifetime Value) por cohorte de usuarios y edad en meses, puedo concluir que:

* **Incremento del LTV con el Tiempo:** Existe un claro patrón de aumento en el `LTV` a medida que los usuarios `maduran dentro de cada cohorte`. Esto indica que, en promedio, los usuarios tienden a `gastar más a medida que permanecen con el servicio por más tiempo.` La cohorte de septiembre de 2017 destaca por tener un salto significativo en el LTV a partir del tercer mes.

```python
# Carga de datos de visitas
visits = pd.read_csv('/datasets/visits_log_us.csv')

# Cambio de tipos de datos para las columnas de tiempo
visits['Start Ts'] = pd.to_datetime(visits['Start Ts'], format="%Y.%m.%d %H:%M:%S")
visits['End Ts'] = pd.to_datetime(visits['End Ts'], format="%Y.%m.%d %H:%M:%S")

# Renombrar columnas para mayor claridad
visits = visits.rename(columns={"Device": "device", "End Ts": "session_end_ts", "Source Id": "source_id", "Start Ts": "session_start_ts", "Uid": "uid"})

# Extracción de mes, semana y fecha de sesión
visits['session_month'] = visits['session_start_ts'].astype('datetime64[M]')
visits['session_week'] = visits['session_start_ts'].astype('datetime64[W]')
visits['session_date'] = visits['session_start_ts'].dt.date

# Encontrar la primera sesión de cada usuario
first_visits = visits.groupby('uid').agg({'session_start_ts': 'min'}).reset_index()
first_visits.columns = ['uid', 'first_session_start_ts']

# Encontrar la primera fecha y mes de sesión
first_visits['first_session_dt'] = first_visits['first_session_start_ts'].dt.date
first_visits['first_session_month'] = first_visits['first_session_start_ts'].astype('datetime64[M]')

# Combinar los datos de las primeras visitas con los datos completos de visitas
visits_full = pd.merge(first_visits, visits, on='uid')

# Calcular la diferencia en meses entre la primera visita y las visitas posteriores
visits_full['age_months'] = ((visits_full['session_month'] - visits_full['first_session_month']) / np.timedelta64(1, 'M')).round().astype('int')

# Carga de datos de órdenes
orders = pd.read_csv('/datasets/orders_log_us.csv')
orders['Buy Ts'] = pd.to_datetime(orders['Buy Ts'])
orders = orders.rename(columns={"Buy Ts": "buy_ts", "Revenue": "revenue", "Uid": "uid"})

# Extracción de día y mes de compra
orders['buy_dt'] = orders['buy_ts'].dt.date
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')

# Encontrar la primera compra de cada usuario
first_orders = orders.groupby('uid').agg({'buy_ts': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_ts']
first_orders['first_order_dt'] = first_orders['first_order_ts'].dt.date
first_orders['first_order_month'] = first_orders['first_order_ts'].astype('datetime64[M]')

# Crear dataframe con las primeras órdenes y las primeras visitas
buyers = pd.merge(first_visits, first_orders, on='uid')
buyers['first_order_dt'] = pd.to_datetime(buyers['first_order_dt'])
buyers['first_session_dt'] = pd.to_datetime(buyers['first_session_dt'])

# Calcular los días hasta la primera compra
buyers['days_to_first_purchase'] = ((buyers['first_order_ts'] - buyers['first_session_start_ts']) / np.timedelta64(1, 'D')).astype('int')

# Carga de datos de costos
costs = pd.read_csv('/datasets/costs_us.csv')
costs['dt'] = pd.to_datetime(costs['dt'])
costs['costs_month'] = costs['dt'].astype('datetime64[M]')

# Calcular el CAC medio
print("Mean CAC is:", round(costs['costs'].sum() / orders['uid'].nunique(), 2))

# CAC por mes
CAC_by_month = pd.merge(costs.groupby('costs_month').agg({'costs': 'sum'}),
                        buyers.groupby('first_order_month').agg({'uid': 'nunique'}),
                        left_index=True,
                        right_index=True)
CAC_by_month['cac'] = CAC_by_month['costs'] / CAC_by_month['uid']

# Gráfico del CAC promedio por mes
CAC_by_month['cac'].plot(figsize=(14, 7), title='Average CAC by month', xlabel='Month', ylabel='c.u.')

# Encontrar la primera fuente por visitante
users = visits_full.sort_values('session_start_ts').groupby('uid').first()
users = users[['source_id']]
buyers = pd.merge(buyers, users, left_on='uid', right_index=True)
buyers_daily = buyers.groupby(['source_id', 'first_order_dt']).agg({'uid': 'count'}).reset_index()
buyers_daily.rename(columns={'uid': 'n_buyers'}, inplace=True)
buyers_daily['first_order_dt'] = pd.to_datetime(buyers_daily['first_order_dt'])

# CAC diario por fuente
costs_ = pd.merge(buyers_daily, costs, left_on=['source_id', 'first_order_dt'], right_on=['source_id', 'dt'])
costs_['cac'] = costs_['costs'] / costs_['n_buyers']

# Gráfico de CAC diario por fuente y mes
costs_.pivot_table(index=['costs_month'], columns='source_id', values='cac', aggfunc='mean').plot(
    figsize=(16, 9), title='CAC by buyer from each source', xlabel='Month', ylabel='c.u.')
plt.show()

# CAC promedio por fuente
cac_by_source = costs_.groupby('source_id').agg({'cac': 'mean'}).round(2)
print(cac_by_source)
```

#### ¿Cuán rentables eran las inversiones? (ROMI)

```python
# Uniendo órdenes con visitas para obtener la fuente de adquisición:
orders_with_source = pd.merge(orders, visits[['uid', 'source_id']], on='uid', how='left')

# Calculando LTV por fuente:
ltv_per_source = orders_with_source.groupby('source_id').agg(total_revenue=('revenue', 'sum')).reset_index()
buyers_per_source = orders_with_source.groupby('source_id').agg(n_buyers=('uid', 'nunique')).reset_index()
ltv_per_source = pd.merge(ltv_per_source, buyers_per_source, on='source_id')
ltv_per_source['LTV'] = ltv_per_source['total_revenue'] / ltv_per_source['n_buyers']

# Utilizar el CAC calculado en tu script anterior:
cac_by_source = costs_.groupby('source_id').agg({'cac': 'mean'}).reset_index()

# Combinar LTV y CAC para cada fuente
performance_df = pd.merge(ltv_per_source[['source_id', 'LTV']], cac_by_source[['source_id', 'cac']], on='source_id', how='left')

# Calcular el ROMI
performance_df['ROMI'] = (performance_df['LTV'] / performance_df['cac'] - 1) * 100

# Redondear y formatear la tabla
performance_df = performance_df.round(2)
performance_df['LTV'] = performance_df['LTV'].apply(lambda x: f"${x:.2f}")
performance_df['cac'] = performance_df['cac'].apply(lambda x: f"${x:.2f}")
performance_df['ROMI'] = performance_df['ROMI'].apply(lambda x: f"{x:.2f}%")

# Mostrar la tabla
print("Tabla con LTV, CAC y ROMI por Fuente de Adquisición:")
display(performance_df.set_index('source_id'))
```
